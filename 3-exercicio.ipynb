```python
import zipfile
import os
import pandas as pd
import plotly.express as px
```

```python
# função que descompacta arquivo
def unzip_file(zip_file_path, extract_path):
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
        extracted_file_name = zip_ref.namelist()
    print("Arquivos extraídos com sucesso! \n Nome do arquivo extraído: \n")
    return extracted_file_name
```

```python
# descompactando o arquivo
zip_file_path = os.getcwd()
extract_path = os.getcwd()
files_names = unzip_file(zip_file_path, extract_path)

for file_name in files_names:
    df = pd.read_csv(os.path.join(extract_path, file_name))
```

```python
# lendo a tabela
df = pd.read_csv(os.path.join(os.getcwd(), 'Stores.csv'), sep=',')
df.head()
```

```python
# renomeando a tabela
df = df.rename(columns={'Store ID':'ID','Store_Area':'Area','Items_Available':'Itens_Disponiveis','Daily_Customer_Count':'Visitantes','Store_Sales':'Vendas'})
df.head()
```

```python
# cria uma nova váriavel que receberá somente as colunas nas quais serão construídos gráficos
subset = df.drop(['ID', 'Area'], axis=1)
subset.describe()
```

```python
# gráfico 1
fig = px.line(df, x='ID', y='Vendas', title='Vendas por ID')
fig.show()
```

```python
# gráfico 2
fig2 = px.bar(df, x='ID', y='Visitantes', title='Visitantes por ID')
fig2.show()
```

```python
# gráfico 3
fig3 = px.area(df, 'ID', 'Itens_Disponiveis', title='Visitantes por ID')
fig3.show()
```
